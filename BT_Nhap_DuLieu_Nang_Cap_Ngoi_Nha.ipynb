```python
import turtle
#Vẽ nhà
t = turtle.Turtle()
t.pencolor("gray")
t.pensize(4)
t.hideturtle() #Làm cho rùa vô hình
t.penup()
t.goto(-250,100)
t.pendown()
t.fillcolor("aquamarine")
t.begin_fill()
t.forward(300)
t.right(90)
t.forward(200)
t.right(90)
t.forward(300)
t.right(90)
t.forward(200)
t.end_fill()

#Vẽ ống khói
t.penup()
t.goto(-80,100)
t.pendown()
t.fillcolor("LightSalmon3")
t.begin_fill()
t.forward(80)
t.right(90)
t.forward(25)
t.right(90)
t.forward(80)
t.right(90)
t.forward(25)
t.end_fill()

#Vẽ mái
t.penup()
t.goto(-250,100)
t.pendown()
t.fillcolor("LightSalmon3")
t.begin_fill()
t.right(135)
t.forward(160)
t.right(90)
t.forward(160)
t.right(135)
t.goto(-250,100)
t.end_fill()

#Vẽ khói
t.penup()
t.goto(-80,200)
t.pendown()
t.goto(-80,230)

t.penup()
t.goto(-65,190)
t.pendown()
t.goto(-65,235)

#Vẽ cửa
t.penup()
t.goto(-150,-100)
t.pendown()
t.fillcolor("LightPink")
t.begin_fill()
t.right(90)
t.forward(130)
t.right(90)
t.forward(100)
t.right(90)
t.forward(130)
t.right(90)
t.forward(100)
t.end_fill()

#Vẽ cây

turtle.done()

```
